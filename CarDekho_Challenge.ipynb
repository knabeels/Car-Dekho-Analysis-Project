```python
f_dataset['Current_Year'] = 2023
```

```python
f_dataset.head()
```

```python
f_dataset['No_Years'] = f_dataset['Current_Year'] - f_dataset['Year']
```

```python
f_dataset.head()
```

```python
f_dataset.drop(columns = ['Year'], inplace = True)
f_dataset.drop(columns = ['Current_Year'], inplace = True)
```

```python
f_dataset
```

```python
f_dataset['Transmission'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
f_dataset['Fuel_Type'] = LE.fit_transform(f_dataset['Fuel_Type'])
f_dataset['Seller_Type'] = LE.fit_transform(f_dataset['Seller_Type'])
f_dataset['Transmission'] = LE.fit_transform(f_dataset['Transmission'])
```

```python
f_dataset
```

```python
f_dataset['Fuel_Type'] = f_dataset['Fuel_Type'].astype('int')
f_dataset['Seller_Type'] = f_dataset['Seller_Type'].astype('int')
f_dataset['Transmission'] = f_dataset['Transmission'].astype('int')
```

```python
f_dataset.corr()
```

```python
f_dataset
```

```python
f_dataset['Fuel_Type'].unique()
```

```python
f_dataset['Seller_Type'].unique()
```

```python
f_dataset['Transmission'].unique()
```

```python
x=f_dataset.iloc[:,1:]
y=f_dataset.iloc[:,0]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train
```

```python
from sklearn import linear_model
Lasso_a = linear_model.Lasso()
```

```python
Lasso_a.fit(x_train, y_train)
```

```python
y_pred = Lasso_a.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred, y_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_pred, y_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred, y_test)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
import pickle
model = open('model.pkl', 'wb')

pickle.dump(Lasso_a, model)
```

```python
pickled_model = pickle.load(open('model.pkl', 'rb'))
pickled_model.predict(x_test)
```
